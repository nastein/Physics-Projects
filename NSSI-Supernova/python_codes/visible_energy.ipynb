```python
import numpy as np
import csv
import os
import matplotlib.pyplot as plt
```

```python
NH = 'NH'
N_directories = ['osc000','osc010','osc001']
IH = 'IH'
I_directories = ['osc100','osc110','osc101']

#Order is es, ibd
SM_NH = [[],[]]
FV_NH = [[],[]]
FP_NH = [[],[]]

SM_IH = [[],[]]
FV_IH = [[],[]]
FP_IH = [[],[]]

SM_NH_err = [[],[]]
FV_NH_err = [[],[]]
FP_NH_err = [[],[]]

SM_IH_err = [[],[]]
FV_IH_err = [[],[]]
FP_IH_err = [[],[]]
```

```python
def ratio_err(a,b,siga,sigb):
  return np.sqrt(float(siga*siga)/float(b*b) + float(sigb*sigb)*(float(a*a)/float(b**4)))
```

```python
for num,d in enumerate(N_directories):

  for subdir, dirs, files in os.walk('/Users/noahsteinberg/Physics/James_Research/NSSI-Supernova/nu_events/'+ d):
    for file in files:
      filepath = subdir + os.sep + file
      if filepath.endswith("1.txt"):
        with open(filepath) as inp:
          reader = csv.reader(inp, delimiter=" ")
          energies = np.array(list(zip(*reader))[1], dtype=float)

          if (num == 0):
            if (filepath.endswith('es1.txt')):
              SM_NH[0].extend(energies)
            if (filepath.endswith('ibd1.txt')):
              SM_NH[1].extend(energies)
          if (num == 1):
            if (filepath.endswith('es1.txt')):
              FV_NH[0].extend(energies)
            if (filepath.endswith('ibd1.txt')):
              FV_NH[1].extend(energies)
          if (num == 2):
            if (filepath.endswith('es1.txt')):
              FP_NH[0].extend(energies)
            if (filepath.endswith('ibd1.txt')):
              FP_NH[1].extend(energies)
```

```python
for num,d in enumerate(I_directories):

  for subdir, dirs, files in os.walk('/Users/noahsteinberg/Physics/James_Research/NSSI-Supernova/nu_events/'+ d):
    for file in files:
      filepath = subdir + os.sep + file
      if filepath.endswith("1.txt"):# and ('es' in filepath):
        with open(filepath) as inp:
          reader = csv.reader(inp, delimiter=" ")
          energies = np.array(list(zip(*reader))[1], dtype=float)

          if (num == 0):
            if (filepath.endswith('es1.txt')):
              SM_IH[0].extend(energies)
            if (filepath.endswith('ibd1.txt')):
              SM_IH[1].extend(energies)
          if (num == 1):
            if (filepath.endswith('es1.txt')):
              FV_IH[0].extend(energies)
            if (filepath.endswith('ibd1.txt')):
              FV_IH[1].extend(energies)
          if (num == 2):
            if (filepath.endswith('es1.txt')):
              FP_IH[0].extend(energies)
            if (filepath.endswith('ibd1.txt')):
              FP_IH[1].extend(energies)
```

```python
bins=range(0,100,3)

for i in range(0,2):
  SM_NH[i],bin_edges = np.histogram(SM_NH[i],bins=bins)
  FV_NH[i],bin_edges = np.histogram(FV_NH[i],bins=bins)
  FP_NH[i],bin_edges = np.histogram(FP_NH[i],bins=bins)
  SM_IH[i],bin_edges = np.histogram(SM_IH[i],bins=bins)
  FV_IH[i],bin_edges = np.histogram(FV_IH[i],bins=bins)
  FP_IH[i],bin_edges = np.histogram(FP_IH[i],bins=bins)

  SM_NH_err[i] = np.sqrt(SM_NH[i])
  SM_IH_err[i] = np.sqrt(SM_IH[i])
  FV_NH_err[i] = np.sqrt(FV_NH[i])
  FV_IH_err[i] = np.sqrt(FV_IH[i])
  FP_NH_err[i] = np.sqrt(FP_NH[i])
  FP_IH_err[i] = np.sqrt(FP_IH[i])
```

```python
fig1 = plt.figure(1,figsize=(18, 16), dpi= 200, facecolor='w', edgecolor='k')

ax1 = fig1.add_subplot(221)
ax1.errorbar(bin_edges[:-1],SM_NH[0], yerr=SM_NH_err[0], fmt='o-',label='Standard ES', markersize=5)
ax1.errorbar(bin_edges[:-1],FP_NH[0], yerr=FP_NH_err[0], fmt='o-',label='g3 = 0.2 ES', markersize=5)
ax1.errorbar(bin_edges[:-1],FV_NH[0], yerr=FV_NH_err[0], fmt='o-',label='g1 = 0.2 ES', markersize=5)
ax1.errorbar(bin_edges[:-1],SM_NH[1], yerr=SM_NH_err[1], fmt='o-',label='Standard IBD', markersize=5)
ax1.errorbar(bin_edges[:-1],FP_NH[1], yerr=FP_NH_err[1], fmt='o-',label='g3 = 0.2 IBD', markersize=5)
ax1.errorbar(bin_edges[:-1],FV_NH[1], yerr=FV_NH_err[1], fmt='o-',label='g1 = 0.2 IBD', markersize=5)


ax3 = fig1.add_subplot(222)
ax3.errorbar(bin_edges[:-1],SM_IH[0], yerr=SM_IH_err[0], fmt='o-',label='Standard ES', markersize=5)
ax3.errorbar(bin_edges[:-1],FP_IH[0], yerr=FP_IH_err[0], fmt='o-',label='g3 = 0.2 ES', markersize=5)
ax3.errorbar(bin_edges[:-1],FV_IH[0], yerr=FV_IH_err[0], fmt='o-',label='FV-NSSI ES', markersize=5)
ax3.errorbar(bin_edges[:-1],SM_IH[1], yerr=SM_IH_err[1], fmt='o-',label='Standard IBD', markersize=5)
ax3.errorbar(bin_edges[:-1],FP_IH[1], yerr=FP_IH_err[1], fmt='o-',label='g3 = 0.2 IBD', markersize=5)
ax3.errorbar(bin_edges[:-1],FV_IH[1], yerr=FV_IH_err[1], fmt='o-',label='g1 = 0.2 IBD', markersize=5)


#xtick = [1000,3500,6000,8500,11000,13500,16000,18000,20000]

ax1.set_ylabel('Events/3 MeV', fontsize = 15)
ax1.set_xlabel('Lepton Energy (MeV)', fontsize = 15)
ax1.set_title('Energy spectrum in NH', y=.95, fontsize = 15)
ax1.legend(loc='upper right');
ax1.set_yscale('log')
ax1.set_xlim(0,61)
for tick in ax1.xaxis.get_major_ticks():
  tick.label.set_fontsize(12)

for tick in ax1.yaxis.get_major_ticks():
  tick.label.set_fontsize(12)

ax3.set_ylabel('Events/3 MeV', fontsize = 15)
ax3.set_xlabel('Lepton Energy (MeV)', fontsize = 15)
ax3.set_title('Energy spectrum in IH', y=.95, fontsize = 15)
ax3.legend(loc='upper right');
ax3.set_yscale('log')
ax3.set_xlim(0,61)
for tick in ax3.xaxis.get_major_ticks():
  tick.label.set_fontsize(12)

for tick in ax3.yaxis.get_major_ticks():
  tick.label.set_fontsize(12)


plt.show()
```
