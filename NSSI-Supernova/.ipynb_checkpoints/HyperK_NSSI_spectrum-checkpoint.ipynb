```python
def make_histo(data,num_bins):
    counts,bin_edges = np.histogram(data,num_bins)
    bin_centers = (bin_edges[:-1] + bin_edges[1:])/2.
    err = np.sqrt(counts)
    return counts, bin_centers, err

def make_histo_ratio(data_x, err_x, data_y, err_y):
    print(data_x)
    print(data_y)
    print(data_x/data_y)
    ratio = data_x/data_y
    err = np.sqrt((err_y/data_y)**2 + (err_x*data_x/(data_y*data_y))**2)
    print(ratio)
    return ratio, err
```

```python
times_es = []
times_ibd = []
times_esg3 = []
times_ibdg3 = []

e_es = []
e_ibd = []
e_esg3 = []
e_ibdg3 = []


lines_es = es.readlines()
lines_ibd = ibd.readlines()
lines_esg3 = esg3.readlines()
lines_ibdg3 = ibdg3.readlines()


for x in lines_es:
    if float(x.split(' ')[0]) > 1000:
        times_es.append(x.split(' ')[0])
        #e_es.append(x.split(' ')[1])
times_es = [float(i) for i in times_es]
#e_es = [float(i) for i in e_es]

print(times_es)
#print(e_es)

for y in lines_ibd:
    if float(y.split(' ')[0]) > 1000:
        times_ibd.append(y.split(' ')[0])
        #e_ibd.append(y.split(' ')[1])
times_ibd = [float(i) for i in times_ibd]
#e_ibd = [float(i) for i in e_ibd]

for z in lines_esg3:
    if float(z.split(' ')[0]) > 1000:
        times_esg3.append(z.split(' ')[0])
        #e_esg3.append(z.split(' ')[1])
times_esg3 = [float(i) for i in times_esg3]
#e_esg3 = [float(i) for i in e_esg3]

for w in lines_ibdg3:
    if float(w.split(' ')[0]) > 1000:
        times_ibdg3.append(w.split(' ')[0])
        #e_ibdg3.append(w.split(' ')[1])
times_ibdg3 = [float(i) for i in times_ibdg3]
#e_ibdg3 = [float(i) for i in e_ibdg3]
```

```python
num_bins =  20

fig=plt.figure(figsize=(18, 16), dpi= 80, facecolor='w', edgecolor='k')

counts_es, bin_centers_es, err_es = make_histo(times_es, num_bins)
counts_ibd, bin_centeres_ibd, err_ibd = make_histo(times_ibd, num_bins)
counts_esg3, bin_centers_esg3, err_esg3 = make_histo(times_esg3, num_bins)
counts_ibdg3, bin_centers_ibdg3, err_ibdg3 = make_histo(times_ibdg3, num_bins)

plt.errorbar(bin_centers_es, counts_es, yerr=err_es, fmt='o', label='elastic scattering')
plt.errorbar(bin_centers_es, counts_ibd, yerr=err_ibd, fmt='x', label='Inverse beta')
plt.errorbar(bin_centers_es, counts_esg3, yerr=err_esg3, fmt='o')
plt.errorbar(bin_centers_es, counts_ibdg3, yerr=err_ibdg3, fmt='x')

#es_ibd_ratio, es_ibd_ratio_err = make_histo_ratio(counts_es, err_es, counts_ibd, err_ibd)
#es_ibd_g3_ratio, es_ibd_g3_ratio_err = make_histo_ratio(counts_esg3, err_esg3, counts_ibdg3, err_ibdg3)

#plt.errorbar(bin_centers_es, es_ibd_ratio, yerr=es_ibd_ratio_err, fmt='o')
#plt.errorbar(bin_centers_esg3, es_ibd_g3_ratio, yerr=es_ibd_g3_ratio, fmt='x')

plt.legend(loc='upper right')
plt.yscale('log')
plt.show()


```
